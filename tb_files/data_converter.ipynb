```python
import numpy as np
import matplotlib.pyplot as plt

in_data = []
nnscores = []
truths = []
trackids = []


NTRACKS = 1000

with open("data/formatted_data.csv", "r") as file:
  line = file.readline()
  for i in range(NTRACKS):
    track = []
    for j in range(5):
      line = file.readline()
      line = line.strip()
      data = line.split(",")
      hit = data[-3:]
      hit = [float(h) for h in hit]
      truth = float(data[3])
      nnscore = float(data[4])
      if(len(track) == 0):
        nnscores.append(nnscore)
        truths.append(truth)
        trackids.append(float(data[0]))
      track.append(hit)
    in_data.append(track)

in_data = np.array(in_data)
nnscores = np.array(nnscores)
truths = np.array(truths)
# print("nnscores:")
# print(nnscores)
# print("truths:")
# print(truths)
# print("in data:")
# print(in_data)
```

```python
labels = ["X","Y","Z"]
from matplotlib.colors import LinearSegmentedColormap
bounds = [(200, 1000+50), (0, 800), (-150, 500)]

def addPlot(ax, index1=0, index2=1):
  ax.scatter(trim_track[:,:,index1], trim_track[:,:,index2], s=15, edgecolor='black', color='pink', alpha=0.1, zorder=2, label="All Data")

  first = "NNScore>0.5"
  c=0
  for t in trim2_track:
    t = np.array([h for h in t if h[0] != 0])
    ax.plot(t[:,index1], t[:,index2], marker='o', ms=np.sqrt(15), markeredgecolor='black', linestyle=':', color='green', alpha=0.05, zorder=4, label=first)
    first = None
    ax.text(t[-1, index1] + 10, t[-1, index2] + 10, "{}".format(trim2_ids[c]), zorder=10, color='red', fontsize=8)
    c+=1
  # print(trim2_NN)


  first = "FPGA returned"
  c=0
  for t in tracks:
    t = np.array([h for h in t if h[0] != 0])
    ax.plot(t[:,index1], t[:,index2], marker='o', ms=np.sqrt(15), markeredgecolor='black', linestyle=':', color='red', alpha=1, zorder=5, label=first)
    first = None
    for h in t:
      ax.add_patch(plt.Circle((h[index1], h[index2]), np.sqrt(3000)/3, edgecolor='black', color='blue', alpha=0.2, zorder=1))
      # ax.add_patch(plt.Circle((h[index1], h[index2]), np.sqrt(3000), edgecolor='black', fill=False, zorder=9))
    ax.text(t[-1, index1] + 10, t[-1, index2] + 10, "{}".format(c), zorder=10, color='blue', fontsize=8)
    c+=1

  ax.grid(True)
  ax.legend()
  ax.set_xlabel(labels[index1])
  ax.set_ylabel(labels[index2])
  ax.set_xlim(bounds[index1])
  ax.set_ylim(bounds[index2])
  ax.set_title("{} vs {}".format(labels[index2], labels[index1]))

def radial(x, y):
  return np.sqrt(x**2 + y**2)

def addRadial(ax):
  ax.scatter(trim_track[:,:,2], radial(trim_track[:,:,0], trim_track[:,:,1]), s=15, edgecolor='black', color='pink', alpha=0.1, zorder=2, label="All Data")


  first = "NNScore>0.5"
  c=0
  for t in trim2_track:
    t = np.array([h for h in t if h[0] != 0])
    ax.plot(t[:,2], radial(t[:,0], t[:,1]), marker='o', ms=np.sqrt(15), markeredgecolor='black', linestyle=':', color='green', alpha=0.05, zorder=4, label=first)
    first = None
    ax.text(t[-1,2], radial(t[-1, 0] + 10, t[-1, 1] + 10), "{}".format(trim2_ids[c]), zorder=10, color='red', fontsize=8)
    c+=1

  first = "FPGA returned"
  c=0
  for t in tracks:
    t = np.array([h for h in t if h[0] != 0])
    ax.plot(t[:,2], radial(t[:,0], t[:,1]), marker='o', ms=np.sqrt(15), markeredgecolor='black', linestyle=':', color='red', alpha=1, zorder=5, label=first)
    first = None
    # for h in t:
    #   ax.add_patch(plt.Circle((h[2], radial(h[0], h[1])), np.sqrt(3000), edgecolor='black', fill=False, zorder=9))
    ax.text(t[-1,2], radial(t[-1, 0] + 10, t[-1, 1] + 10), "{}".format(c), zorder=10, color='blue', fontsize=8)
    c+=1

  # print(trim2_NN)
  ax.grid(True)
  ax.legend()
  ax.set_xlabel("z")
  ax.set_ylabel("r")
  # ax.set_xlim()
  ax.set_ylim(200, 1000+50)
  ax.set_title("Radial Vs Z")
```
