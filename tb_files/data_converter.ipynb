```python
import numpy as np
import matplotlib.pyplot as plt

in_data = []
nnscores = []
truths = []
trackids = []


NTRACKS = 100

with open("data/formatted_data.csv", "r") as file:
  line = file.readline()
  for i in range(NTRACKS):
    track = []
    for j in range(5):
      line = file.readline()
      line = line.strip()
      data = line.split(",")
      hit = data[-3:]
      hit = [float(h) for h in hit]
      truth = float(data[3])
      nnscore = float(data[4])
      if(len(track) == 0):
        nnscores.append(nnscore)
        truths.append(truth)
        trackids.append(float(data[0]))
      track.append(hit)
    in_data.append(track)

in_data = np.array(in_data)
nnscores = np.array(nnscores)
truths = np.array(truths)
# print("nnscores:")
# print(nnscores)
# print("truths:")
# print(truths)
# print("in data:")
# print(in_data)
```
