```python
import numpy as np

full_in_data = []

with open("data/tb_track_data.dat", "r") as file:
  for i in range(100):
    line = file.readline()
    line = line.strip()
    data = line.split(" ")
    # print(data)
    data = [float(d) for d in data]
    # data = np.array(data)
    full_in_data.append(data)

full_in_data = np.array(full_in_data)
print("full in data:")
print(full_in_data)
  
out_data = []

with open("data/tb_NNscore.dat", "r") as file:
  for i in range(100):
    line = file.readline()
    line = line.strip()
    data = line.split(" ")
    # print(data)
    data = [float(d) for d in data]
    data = data[0]
    # data = np.array(data)
    out_data.append(data)

out_data = np.array(out_data)
print("\nout data:")
```
