```python
import numpy as np

full_in_data = []

with open("tb_track_data.dat", "r") as file:
  for i in range(100):
    line = file.readline()
    line = line.strip()
    data = line.split(" ")
    # print(data)
    data = [float(d) for d in data]
    # data = np.array(data)
    full_in_data.append(data)

full_in_data = np.array(full_in_data)
print("full in data:")
print(full_in_data)
  



out_data = []

with open("tb_NNscore.dat", "r") as file:
  for i in range(100):
    line = file.readline()
    line = line.strip()
    data = line.split(" ")
    # print(data)
    data = [float(d) for d in data]
    data = data[0]
    # data = np.array(data)
    out_data.append(data)

out_data = np.array(out_data)
print("\nout data:")
print(out_data)
```

```python
import matplotlib.pyplot as plt

in_data = []
NHITS = 5

for d in range(len(full_in_data)):
  # data = [full_in_data[d][i:i+3] for i in range(0, len(full_in_data[d]), 3)]
  data = []
  for i in range(NHITS):
    x = full_in_data[d][i + 0 * NHITS]
    y = full_in_data[d][i + 1 * NHITS]
    z = full_in_data[d][i + 2 * NHITS]
    data.append(np.array([x,y,z]))
  data = np.array(data)
  in_data.append(data)
in_data = np.array(in_data)
print(in_data.shape)
print(in_data)

# plt.scatter(in_data[0, ])
```

```python
with open("tb_output.dat", 'r') as file:
  data = file.read().split("\n")[:-1]

trks = [data[d:d+5+1] for d in range(0,len(data), 5+1)]
print(trks)

nn_score = [d[0] for d in trks]
trks = [d[1:] for d in trks]

print(nn_score)
print(trks)

tracks = []
for trk in trks:
  track = []
  for hit in trk:
    hits = hit.split(" ")
    # hits = [int(h) for h in hits]
    hits = [float(h) for h in hits]
    track.append(hits)
  track = np.array(track)
  tracks.append(track)
tracks = np.array(tracks)

print(tracks.shape)
print(tracks)


```

```python
labels = ["X","Y","Z"]
from matplotlib.colors import LinearSegmentedColormap
bounds = [(200, 1000+50), (0, 800), (-150, 500)]

def addPlot(ax, index1=0, index2=1):
  c = 0
  ax.scatter(trim_track[:,:,index1], trim_track[:,:,index2], s=15, edgecolor='black', color='pink', alpha=0.1, zorder=2, label="All Data")
  first = "FPGA returned"
  for t in tracks:
    t = np.array([h for h in t if h[0] != 0])
    ax.plot(t[:,index1], t[:,index2], marker='o', ms=np.sqrt(15), markeredgecolor='black', linestyle=':', color='red', alpha=1, zorder=5, label=first)
    first = None
    for h in t:
      ax.add_patch(plt.Circle((h[index1], h[index2]), np.sqrt(3000)/3, edgecolor='black', color='blue', alpha=0.2, zorder=1))
      # ax.add_patch(plt.Circle((h[index1], h[index2]), np.sqrt(3000), edgecolor='black', fill=False, zorder=9))

    # bnds = [bounds[index1][0], bounds[index1][1], bounds[index2][0], bounds[index2][1]]
    # for h in t:
    #   x, y = np.meshgrid(np.linspace(bnds[0], bnds[1], 500), np.linspace(bnds[2], bnds[3], 500))
    #   r = np.sqrt((x - h[index1]) ** 2 + (y - h[index2]) ** 2)
    #   # z = np.where(r < np.sqrt(3000), np.nan, np.clip(np.sqrt(6000) - r, 0, np.inf))
    #   z = np.clip(1/r, 0, 1/15)
    #   cmap = LinearSegmentedColormap.from_list('', ['#FFFFFF00', 'red'])
    #   # cmap = LinearSegmentedColormap.from_list('', ['red', '#FFFFFF00'])
    #   # cmap.set_bad('#ff334466')
    #   ax.imshow(z, cmap=cmap, extent=bnds, origin='lower', zorder=3)

    ax.text(t[-1, index1] + 10, t[-1, index2] + 10, "{}".format(c), zorder=10, color='blue', fontsize=8)
    c+=1

  first = "NNScore>0.5"
  c = 0
  for t in trim2_track:
    t = np.array([h for h in t if h[0] != 0])
    ax.plot(t[:,index1], t[:,index2], marker='o', ms=np.sqrt(15), markeredgecolor='black', linestyle=':', color='green', alpha=0.7, zorder=4, label=first)
    first = None
    ax.text(t[-1, index1] + 10, t[-1, index2] + 10, "{}".format(trim2_ids[c]), zorder=10, color='red', fontsize=8)
    c+=1
  # print(trim2_NN)
  ax.grid(True)
  ax.legend()
  ax.set_xlabel(labels[index1])
  ax.set_ylabel(labels[index2])
  ax.set_xlim(bounds[index1])
  ax.set_ylim(bounds[index2])
  ax.set_title("{} vs {}".format(labels[index2], labels[index1]))

def radial(x, y):
  return np.sqrt(x**2 + y**2)

def addRadial(ax):
  c = 0
  ax.scatter(trim_track[:,:,2], radial(trim_track[:,:,0], trim_track[:,:,1]), s=15, edgecolor='black', color='pink', alpha=0.1, zorder=2, label="All Data")
  first = "FPGA returned"
  for t in tracks:
    t = np.array([h for h in t if h[0] != 0])
    ax.plot(t[:,2], radial(t[:,0], t[:,1]), marker='o', ms=np.sqrt(15), markeredgecolor='black', linestyle=':', color='red', alpha=1, zorder=5, label=first)
    first = None
    # for h in t:
    #   ax.add_patch(plt.Circle((h[2], radial(h[0], h[1])), np.sqrt(3000), edgecolor='black', fill=False, zorder=9))
    ax.text(t[-1,2], radial(t[-1, 0] + 10, t[-1, 1] + 10), "{}".format(c), zorder=10, color='blue', fontsize=8)
    c+=1

  first = "NNScore>0.5"
  c=0
  for t in trim2_track:
    t = np.array([h for h in t if h[0] != 0])
    ax.plot(t[:,2], radial(t[:,0], t[:,1]), marker='o', ms=np.sqrt(15), markeredgecolor='black', linestyle=':', color='green', alpha=0.7, zorder=4, label=first)
    first = None
    ax.text(t[-1,2], radial(t[-1, 0] + 10, t[-1, 1] + 10), "{}".format(trim2_ids[c]), zorder=10, color='red', fontsize=8)
    c+=1

  # print(trim2_NN)
  ax.grid(True)
  ax.legend()
  ax.set_xlabel("z")
  ax.set_ylabel("r")
  # ax.set_xlim()
  ax.set_ylim(200, 1000+50)
  ax.set_title("Radial Vs Z")
```
